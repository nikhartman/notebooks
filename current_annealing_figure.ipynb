```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import gridspec
mpl.style.use('/Users/nik/Dropbox/Notebooks/thesisrc.py')
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import pandas as pd
import lvdatatools as lvdt
```

```python
data_dir = '/Volumes/JUPITER/Data/test'
```

```python
import os, time
from datetime import datetime

# look for all *.txt
devices = []
for dirpath, subdirs, files in os.walk(data_dir):
    for f in files:
        if f.endswith('.bin'):
            devices.append(os.path.join(dirpath, f))
devices.sort(key=lambda n: os.path.getmtime(n))

dates = []
for d in devices[::-1]:
    #date = time.localtime(os.path.getmtime(d))
    date = datetime.fromtimestamp(os.path.getmtime(d))
    dates.append(date)
    print '{0} -- {1}'.format(d[27:], datetime.strftime(date, '%m/%d/%Y'))
```

```python
# anneal_scfmf8_b.txt


f = os.path.join(data_dir, 'anneal_scfmf8_b.bin')
df = lvdt.get_data_2d(f)

fig, ax = plt.subplots()
ax.plot(df['Gate'], df['Iavg']*1e9, lw=3)
ax.set_ylim(-250, 100)
ax.set_xlim(-5,5)
ax.set_xlabel('$V_{bias} (V)$', fontsize = 24)
ax.set_ylabel('I (nA)', fontsize = 24)

fig.savefig('/Users/nik/Dropbox/Repos/thesis/figures/chapter4/current_anneal.pdf')
```
