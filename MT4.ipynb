### R-T Data

```python
data_dir = '/Volumes/EDDropbox/Data/MT4/R-T/'
for f in glob.glob(os.path.join(data_dir, '*.txt')):
    print f.split('/')[-1]
```

```python
# f = os.path.join(data_dir, 'mt4_cooling_17_6_0.txt')
# fo = open(f)
# for line in fo:
#     print line
    
# [T Cond ...]

for f in glob.glob(os.path.join(data_dir, '*.txt')):
    data = np.loadtxt(f)
    T = data[:,0]
    cond = data[:,1]/3.875e-5

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(T, cond)
    ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
    ax.set_xlabel(r'$T (K)$')
    ax.set_ylabel(r'$G (\frac{e^2}{h})$')
```

```python
# matrix 6

f = os.path.join(data_dir, 'mt4_13-20_4K_matrix_3_Vdc_0.03-0.03_step_0.001mV.txt')

df = pd.read_csv(f, sep=None)
bias = df.ix[1:,0].astype(np.float)*1e3
gate = df.columns[1:].astype(np.float)
cond = df.ix[1:,1:].values/3.875e-5

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
limits = [gate.min(), gate.max(), bias.min(), bias.max()]
im = ax.imshow(cond, extent = limits, 
               cmap = plt.cm.seismic, #vmin = -0.6, vmax = 1.0, 
               origin = 'lower', aspect = 'auto', interpolation = 'nearest')
ax.set_xlabel('Gate (V)')
ax.set_ylabel('Bias (mV)') 
ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
cb = plt.colorbar(im)
cb.set_label(r'$G (\frac{e^2}{h})$')
```

Does not look like the bias is working correctly here.
