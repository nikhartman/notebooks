```python
data_dir = '/Volumes/EDDropbox/Data/AuD2/Gp_V/'
os.listdir(data_dir)
```

```python
f = os.path.join(data_dir, 'Gp_V_11-9_4p4K_0T_3.txt')
# fo = open(f)
# for line in fo:
#     print line

data = np.loadtxt(f)
gate = data[:,0]
r = data[:,1]/1000 # in kOhms
g = data[:,2]/3.875e-5 # in e^/h

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(gate, g)
ax.set_xlabel(r'$V_g (V)$')
ax.set_ylabel(r'$G (\frac{e^2}{h})$')
ax.set_ylim(0.0204, 0.0206)
```
