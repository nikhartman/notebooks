Looking at the first high resolution measurement I did.

```python
df = lvdt.vi_gateswp_data('VI-gate_17-19_13_f.bin')
df = df.groupby(df.index).mean()
df.columns.name = 'gate'
df.index.name = 'bias'
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
limits = [df.columns[0], df.columns[-1],df.index[0],df.index[-1]]
im = ax.imshow(df*1e7+0.0, extent = limits, 
               cmap = plt.cm.seismic,
               origin = 'lower', aspect = 'auto', interpolation = 'None')
ax.set_xlabel(r'$Gate (V)$', fontsize=16)
ax.set_ylabel(r'$Bias (mV)$', fontsize=16)
ax.set_title('VI-gate_17-19_13_f.bin'+', B=0T')
cb = plt.colorbar(im)
cb.set_label(r'$I (nA)$', fontsize=16)
```

```python
fig.savefig('VI-gate_17-19_13_f'+'.png', dpi=200)
```

```python
cond = df.diff()
cond.max().max()
cond = cond/cond.max().max()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
limits = [cond.columns[0], cond.columns[-1],cond.index[0],cond.index[-1]]
im = ax.imshow(cond, extent = limits, 
               cmap = plt.cm.seismic, vmin = -0.05, vmax = 0.07, 
               origin = 'lower', aspect = 'auto', interpolation = 'None')
ax.set_xlabel(r'$V_{BG} (V)$', fontsize=20)
ax.set_ylabel(r'$V_{bias} (mV)$', fontsize=20)
ax.set_title('VI-gate_17-19_13_f.bin'+', B=0T')
cb = plt.colorbar(im)
cb.set_label(r'$G (arb)$', fontsize=20)
```

```python
fig.savefig('VI-gate_17-19_13_f'+'_cond.png', dpi=200)
```

Let's see if averaging over a few gate values changes much. And look at that first diamond near 0 gate voltage.

```python
cond_roll = pd.rolling_mean(cond, 2, axis=1)
cond_roll = pd.rolling_mean(cond_roll, 2, axis=0)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,1,1)
limits = [0, 0.135,cond.index[0],cond.index[-1]]
im = ax.imshow(cond_roll.loc[limits[2]:limits[3],limits[0]:limits[1]], extent = limits, 
               cmap = plt.cm.seismic, vmin = -0.05, vmax = 0.07, 
               origin = 'lower', aspect = 'auto', interpolation = 'none')
ax.set_xlabel(r'$V_{BG} (V)$', fontsize=20)
ax.set_ylabel(r'$V_{bias} (mV)$', fontsize=20)
ax.set_title('VI-gate_17-19_13_f.bin'+', B=0T')
cb = plt.colorbar(im)
cb.set_label(r'$G (arb)$', fontsize=20)
```
