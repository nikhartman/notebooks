### Cond vs Gate (AC)

```python
filelist = ['mt7_9-11_0T_AConly_10V-10m6div.txt',
'mt7_9-11_0T_AConly_1V-10m5div.txt',
'mt7_9-11_0T_gate-sweep_n1.txt',
'mt7_9-11_0T_gate-sweep_n2.txt',
'mt7_9-11_0T_gate-sweep_n2_1.txt',
'mt7_9-11_0T_gate-sweep_n2_570Hz.txt',
'mt7_9-11_0T_gate-sweep_n2_570Hz_backward.txt']
filelist = [os.path.join(data_dir, f) for f in filelist]

bias = np.array([0, 0, -1, -2, -2, -2, -2], dtype=np.float)

for f, b in zip(filelist, bias):
    data = np.loadtxt(f, delimiter='\t')
    gate = data[:,0]
    cond = data[:,1]/3.875e-5

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(gate, cond, label='bias = {0}mV'.format(b))
    ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
    ax.set_xlabel(r'$V_g (V)$')
    ax.set_ylabel(r'$G (\frac{e^2}{h})$')
    ax.legend(fontsize = 16)
    #ax.set_ylim(0,None)
```

```python
data_dir = '/Volumes/EDDropbox/Data/MT7/GpV_MATRIX/'
for i, f in enumerate(glob.glob(os.path.join(data_dir, '*.txt'))):
    print i, f.split('/')[-1]
```

```python
# 1, 2, 3, 7, 9, 10, 11, 12
filelist = [
'mt7_9-11_4p2K_0T_matrix_g_Vdc_-2.502.50_step_0.050mV.txt',
'mt7_9-11_4p2K_0T_matrix_c_Vdc_-10.0010.00_step_0.250mV.txt',
'mt7_9-11_4p2K_0T_matrix_h_Vdc_-2.502.50_step_0.040mV.txt',
'mt7_9-11_4p2K_0T_matrix_d_using-vdriveadder_Vdc_-10.0010.00_step_0.200mV.txt',
'mt7_7-9_4p2K_0T_matrix_a_Vdc_-10.0010.00_step_0.250mV.txt',
'mt7_10-13_4p2K_0T_matrix_a_Vdc_-5.005.00_step_0.100mV.txt',
'mt7_9-11_4p2K_0T_matrix_i_Vdc_-6.006.00_step_0.100mV.txt',
'mt7_9-11_4p2K_1T_matrix_h_Vdc_-2.502.50_step_0.040mV.txt'
]
filelist = [os.path.join(data_dir, f) for f in filelist]

for f in filelist:
    df = pd.read_csv(f, sep='\t')
    bias = df.iloc[1:,0].astype(np.float)
    gate = df.columns[1:].astype(np.float)
    cond = df.iloc[1:,1:]/3.875e-5
    limits = [gate.min(), gate.max(), bias.min(), bias.max()]

    fig = plt.figure(figsize=(16,8))
    ax = fig.add_subplot(1,1,1)
    im = ax.imshow(cond, extent = limits, 
                   cmap = plt.cm.seismic, #vmin = -0.6, vmax = 1.0, 
                   origin = 'lower', aspect = 'auto', interpolation = 'nearest')
    ax.set_xlabel('Gate (V)')
    ax.set_ylabel('Bias (mV)') 
    ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
    cb = plt.colorbar(im)
    cb.set_label(r'$G (\frac{e^2}{h})$')
```
