Let's see what combinations of two times, including the original section time (so I don't have to search for open rooms), can accomodate everyone.

```python
for i, col in poll.iteritems():
    cross = np.logical_or(poll['T1330'], col)
    if cross.sum() > 10:
        print 'at t1 = {} or t2 = {}, {} of {} people can meet' \
              .format('T1330' , col.name, cross.sum(), len(names))
```

No combination of Tuesday at 1:30pm and another time can accommodate all the respondants.

Let's try any combination of two times.

```python
for i, col in poll.iteritems():
    for j, dol in poll.iteritems():
        if j < i:
            cross = np.logical_or(col, dol)
            if cross.sum() > 14:
                print 'at t1 = {} or t2 = {}, {} of {} people can meet' \
                      .format(col.name , dol.name, cross.sum(), len(names))
```
