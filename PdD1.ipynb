### Cond vs Gate

```python
data_dir = '/Volumes/EDDropbox/Data/PdD1/cond_gate/'
for i, f in enumerate(glob.glob(os.path.join(data_dir, '*.txt'))):
    print i, f.split('/')[-1]
```

```python
for f in glob.glob(os.path.join(data_dir, '*.txt')):
    data = np.loadtxt(f, delimiter='\t')
    # [Gate R G ? ? T ?]
    gate = data[:,0]
    cond = data[:,2]/3.875e-5

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(gate, cond)
    ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
    ax.set_xlabel(r'$V_g (V)$')
    ax.set_ylabel(r'$G (\frac{e^2}{h})$')
```

### Bias/Gate Sweep

```python
data_dir = '/Volumes/EDDropbox/Data/PdD1/cond-matrix/'
for i, f in enumerate(glob.glob(os.path.join(data_dir, '*.txt'))):
    print i, f.split('/')[-1]
```

```python
# 0 pdd1_7-9_4p2K_0T_c_Vdc_-6.006.00_step_0.010mV.txt
f = os.path.join(data_dir, 'pdd1_7-9_4p2K_0T_c_Vdc_-6.006.00_step_0.010mV.txt')

df = pd.read_csv(f, sep='\t')
bias = df.columns[2:].astype(np.float)
gate = df['B-gate']
cond = df.iloc[:,2:].transpose()/3.875e-5
limits = [gate.min(), gate.max(), bias.min(), bias.max()]

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
im = ax.imshow(cond, extent = limits, 
               cmap = plt.cm.seismic, #vmin = -0.6, vmax = 1.0, 
               origin = 'lower', aspect = 'auto', interpolation = 'nearest')
ax.set_xlabel('Gate (V)')
ax.set_ylabel('Bias (mV)') 
ax.set_title(r'$\verb|{0}|$'.format(f.split('/')[-1]))
cb = plt.colorbar(im)
cb.set_label(r'$G (\frac{e^2}{h})$')
```
