```python
thesis_dir = '/Users/nik/Dropbox/Repos/thesis/figures/fmdots/'
Image(filename=os.path.join(thesis_dir, 'scf75_energy_calc.png'), width=600)
```

Looking at the levels on the left, there seems to be a missing peak that should cut through zero. With that in mind, I'm claiming the first peak seen is the first excited state. This gives a level spacing of $\Delta E \sim 3.5meV$, which corresponds to a dot size $L \sim 140nm$.

That is why this measurement was singled out to discuss this model. It shows a level spacing about 2x larger than the other devices measured, which is good evidence for more than one quantum dot in series.

```python
Image(filename=os.path.join(thesis_dir, 'scf75_gatesweep.png'), width=600)
```

The irregular spacing in as a function of the gate also suggests that there is more than just a single quantum dot. If the spacing were even, without showing four fold symmetry, it might be explained by the charging energy being much larger than the level spacing. That does not seem to be the case here, the peak spacing is irregular and does not show any sensible symmetry.
