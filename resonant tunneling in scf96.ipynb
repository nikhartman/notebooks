This sample shows the resonant tunneling behavior at one small range of gate voltages. Here are some plots and a little bit of analysis about it. The sample is named SCF96, all quantum dots are designed to be 300nm in length, and the leads were made of ebeam evaporated cobalt.

```python
import os
from IPython.display import Image 

fig_dir = '/Volumes/JUPITER/Data/SCF96/figures/'
thesis_dir = '/Users/nik/Dropbox/Repos/thesis/figures/fmdots/'
Image(filename=os.path.join(fig_dir, 'scf96_q3.jpg'), width=600)
```

The leads being measured are labelled 16 (narrow) and 17 (wide) in the image. The two probe resistance at room temperature was $R = 400k\Omega$. A measurement on a different pair of leads shows that this nanotube is semiconducting.

From the conductance data, the dot size can be calculated using $\Delta E = \frac{0.5eV}{L(nm)} = 2.0meV$, which gives $L = 250nm$. Again, the dot should be 300nm, so this seems fine. 

```python
Image(filename=os.path.join(thesis_dir,'scf96_resonant_tunneling_conductance.png'), width=600, height=900)
```

The peak at $V_{gate} = 20V$ in the 0T data is where the level splitting was measured. Ignoring the negative differential conductance peaks in the 2T data, it is clear that the resonant tunneling features are suppressed in a finite field.

Looking at some cuts through the conductance plot gives an idea of the energy scales for this tunneling process.

```python
Image(filename=os.path.join(thesis_dir,'scf96_resonant_tunneling_cuts.png'), width=600, height=900)
```

These cuts at fixed gate voltage clearly show there are three positive conductance peaks each followed by a matching negative conductance peak. the first peak shows up at $\Delta E \sim 8meV$, the second at $\Delta E \sim 13meV$, and the third at $\Delta E \sim 22meV$. The splitting between the positive and negative conductance peaks is $\sim 2meV$. All of these values have some gate voltage dependance, but for now it seems more important just to get an idea of the energy scales. 

The features appear in a range of $V_{gate}$ from 10 to 20V. Using the value for the lever arm in this sample $\alpha_G = 0.006$ and $U = \alpha_G e V_{gate}$, this range of gate voltages is found to correspond to energies from 60 to 120meV.

Looking at the corresponding 1/f noise amplitude actually gives some proof that there are 3 pairs of levels, not 3 single levels with odd peak shapes. In the noise data both the positive and negative conductance peaks appear as increases in the noise amplitude. 

```python
Image(filename=os.path.join(thesis_dir,'scf96_resonant_tunneling_noise-cuts.png'), width=600, height=900)
```

I'm not totally convinced this is a valid way of using the noise data, but it does look nice here. It seems to suggest that the tunnel rates onto the dot are the same for the positive and negative conductance peaks.

This sample is tilted about 45$^\circ$ in the field, but it does show some features in the magnetoresistance at low fields.

```python
Image(filename=os.path.join(thesis_dir,'scf96_bsweep.png'), width=600)
```

This was taken at $V_{gate}=7V$ and $V_{bias}=20mV$, which should be outside of the blockaded region. The effect has a magnitude of about -1.5% and occurs at about 150mT. A previous sample showed that leads of this size have a coerrsive field of about 80mT and had a TMR of -2%. Nievely multiplying $\sqrt{2}*80mT$ (because the field is 45$^\circ$ off of the easy axis) says this feature might be due to the magnetization of the leads in the field.
