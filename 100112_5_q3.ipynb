```python
data_dir = '/Volumes/JUPITER/Data/100112_5_q3/room_temp/'

for f in glob.glob(os.path.join(data_dir, '*.dat')):
    data = np.loadtxt(f)
    gate = data[:,0]
    i = data[:,1]*1e6

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(gate, i)
#     ax.set_title(f.split('/')[-1])
    ax.set_xlabel(r'$V_{gate} (V)$')
    ax.set_ylabel(r'I ($\mu$A)')
```

```python
data_dir = '/Volumes/EDDropbox/Data/100112_5_q3/iv_DAQgate/'

for f in glob.glob(os.path.join(data_dir, '*.dat')):
    data = np.loadtxt(f)
    bias = data[0,1:]*1e3
    gate = data[1:-1,0]
    i = np.rot90(data[1:-1,1:]*1e9)

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ext = [gate.min(), gate.max(), bias.min(), bias.max()]
    ax.imshow(i, interpolation = 'nearest', aspect = 'auto', cmap = plt.cm.jet, extent = ext)
    ax.set_title(f.split('/')[-1])
    ax.set_xlabel(r'$V_g (V)$')
    ax.set_ylabel(r'$V_b (mV)$')
```
